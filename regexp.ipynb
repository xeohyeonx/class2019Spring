위 cell은 이렇게 불러온 단어장을 list로 해서 그걸 for loop 돌리는 것
in 뒤의 내용을 먼제 읽어보면 이게 우리가 전에 불러온 단어장 list이고, 이 단어장 전체의 항목만큼 for loop를 돌려 w에 넣는데, 이때 조건으로 if 뒤에 제시된 내용을 보면 w.islower()를 통해 for loop를 돌린 w가 소문자일때만 w로 쌓아라 라고 되어있으니 결과로는 소문자만 wordlist로 쌓이게됨. (이 wordlist도 당연히 type는 list) 

위에서 예시로 설정한 변수인 dic을 이용해서 표현하려면 in 다음 부분을 dic로 해서 [w for w in dic if w. islower()] 이렇게 쓰면 됨. 

=in 다음에 list가 있고 그 list에는 항목이 있을텐데 그 항목 수 만큼 for loop 가 돌 것이고, 돌면서 w를 하나하나 타고 들어가는데 이게 소문자인지 대문자인지 물어서 소문자이면 그걸 list로 쌓는것. 그럼 wordlist가 소문자로 정제되어서 소문자 단어만 골라서 나올 것이고, 그 중 10개만 보는기 위해 [:10]라고 범위 설정.

```python
for w in nltk.corpus.words.words.('en') if w.islower()
    print(w)
```

이것은 위에 꺼랑 똑같은데, 이것은 빈 리스트인 a를 하나 만들고 for loop를 돌린 다음에 if w.islower()이면 빈 리스트에 넣어라 라는 말을 자세히 쓴 버전이고 위에 wordlist가 그걸 짧게 작성한 것

```python
re.search('abc', 'abcde')
```

위 함수의 입력은 'abc'와 'abcde' 이거 두개인데, 첫번째 항목은 re.search에 대한 조건, 두번째가 찾는 대상을 뜻함. 그러니까 abcde라는 string에서 abc를 찾으라는 말.  
(abcde와 abc의 순서가 거꾸로 되면 실행이 안됨. 더 작은 것에서 더 큰걸 찾으라는 말이니까) 

```python
result = [w for w in wordlist if re.search('ed$', w)][:10]
print(result[:10])
```

word list는 string 들의 list인데, 이를 re.search 하려면 re.search는 string에만 되니까 forloop를 사용해서 하나하나의 string 에 적용시켜야함. 
그걸 한꺼번에 한줄에 하는게 위의 cell인데, 우선 in 다음의 내용을 보면 wordlist 갯수만큼 loop를 돌려 그를 w로 가져가는데, if 뒤의 조건에 따라 w에 들어감. 
그 조건에 의해 string의 끝이 ed로 끝나면 찾아서 그걸 result로 차곡차곡 쌓으라는 뜻.
(if 는 뭐라도 찾아지면 통과. if가 적용되기 전의 w랑 비교를 했을때 ed로 끝나는게 찾아지면 w에 쌓이는것.) 
그다음의 [:10]는 list로 했지만 그 중 10개 까지만 result로 받는것.
그럼 밑의 줄의 [:10]은 redundant하기 때문에 빼도됨. 

(if 부분이 없으면 for loop 돌리고 그냥 그걸 다 조건없이 w 에 넣는 것이기 때문에 그럼 w=wordlist이니까 for loop를 하는 의미가 없음) 

사실 ^과 $는 굳이 안써도 되긴 하지만 순서를 정확히 명시하는 의미가 있음. 예를 들어 ^없이 내용이 있고 $만 있으면 뒤가 내용과 같은 것을 찾으면되고, 그 반대의 경우는 첫부분이 그러한 요소를 찾으면 되는 것.

```python
result = [w for w in wordlist if re.search('^..j..t..$', w)][:10]
print(result[:10])
```

아까도 확인했듯이 regular expression에서의 달러 표시는 string의 끝을 의미하고, "."은 지정되지 않은 하나의 character, ^는 string의 시작을 의미함. 그럼 ..j..t.. 이렇게 총 길이가 8개인 게 무조건 받아진다는 게 조건의 내용.  
여기서도 [:10]는 redundant 하니까 윗 줄의 그것은 지우는게 나음. 

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)] 
print(result[:10])
```

여기도 ^이랑 달러가 있으니까 처음과 끝이 정해져있고, regular expression에 []가 나오면 그 안의 요소 중 하나(무엇이 나와도 상관이 없음)라는 뜻.  
근데 이게 하나씩 최소한 나와서 총 4개의 character가 있는게 되는지 gggooo이렇게 나와도 되는지 이런게 헷갈리면 run을 해서 결과를 보면 각 barcket에서 하나씩을 의미한다는 것을 알 수 있음.
그러니까 영어 소문자 단어 중에서 첫번째 알파벳은 ghi 중 하나, 두번째는 mno 중 하나.. 이런 식의 조건에 맞는 것을 찾는 게 위 cell의 의미. 


```python
result = [w for w in wordlist if re.search('^[ah]+$', w)][:10]
print(result[:10])
```

string의 처음과 끝은 정해졌고, +은 one or more 이라는 뜻.(그니까 반복해서 나올수도 있다는 뜻) 
ah중의 하나이고 그게 1회 이상 매치되는것은 그럼 a, aa, ah aha h ha hah 이런식의 결과를 의미함. 

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
```

이건 이제 여태 사용했던 것과는 또다른 corpus를 불러오는 것. 이를 실행시키면 이것도 사전은 사전인데 아까와는 다르게 생긴 사전이 나옴. wsj=worldstreetjournal이라는 뜻이고 이것의 variable도 list. 
=>그래서 바로 regular expression은 안되고 for loop로 하나하나 적용시켜야. 

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
```

이걸 보면 처음과 끝은 명시가 되어있고, [0-9]+는 []안의 것 중 하나가 아닌 하나 이상이 나온다는 뜻. 그 뒤의 "." 는 \로 인해 그냥 아무 character가 아니라 진짜 말 그대로의 . 을 의미함.
그럼 string의 처음에 숫자가 여러 개 나올 수 있고 그 다음에 . 이 나와야하고 그 다음에도 또 아무리 많아도 여러개의 숫자가 나와도 된다라는 말이니까 이것은 결국 실수를 찾으라는 말

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
print(wordlist[:10])
```

이렇게 하면 위에서 숫자의 경우와 비슷하게 알파벳이 뭐가 나오든 상관없이 하나 이상 나와야하고, string의 끝을 의미하는 달러사인 앞에 \$가 있으니까 이는 맨 앞에는 알파벳 하나이상 + 제일 끝에 달러사인이 붙는 걸 찾으라는 뜻. 

(다시 정리하자면 그냥 []는 그 안의 내용 중 오직 하나를 의미. 그래서 [A-Z0-9]하면 알파벳이든 숫자든 하나만 나오는거.
A-Z가 대문자로 되어있으니까 소문자는 안나옴. A-Za-z이라고 해야 대문자 소문자 상관없이 아무 알파벳이라는 말) 

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
```

worldstreetjournal의 중에서 조건에 부합하는 걸 w로 쌓아서 wordlist를 만들고 그것의 결과를 보는 cell.
{n}에서의 n은 integer, 즉 자연수고, 이건 몇번 반복하는지를 알려주는 숫자이기 때문에 위 조건의 0-9의 숫자 중 하나를 고르는 것을 4번 반복하는 것= four digit숫자를 찾으라는 말. 
(근데 왜 단어장에 저렇게 들어가있냐? 하면 그냥 뭐 corpus 기반으로 찾았으니까,,,,연도일 가능성이 높음)

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
```

[0-9]+면 0에서 9 중 하나 이상 (숫자가 하나이상)이고 -는 "-"라는 문자가 나와야된다는것. 그리고 소문자 알파벳이 {}번 만큼 나와야되므로 3개에서 5개정도 나와야한다는 말. 
그럼 결과로 막 10-day, 10-lap, 10-year 등이 나옴. 

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

{5,}면 5개 이상이다 라는 말이고, 그럼 소문자 알파벳 5개 이상, "-", 또 소문자 알파벳이 2-3개, 또 -, 그리고 소문자 알파벳이 6개 이하로 나오는걸 찾아라 라는 말.

```python
wordlist = [w for w in wsj if re.search('(ed|ing)$', w)]
print(wordlist[:10])
```

이러면 리스트가 나오는데, 위 cell은 findall로 두개 이상의 모음 combination이 있으면 그걸 vs로 해서 쭉 list up 하라는 의미이기 때문.

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```
